```python
#link = 'https://indeed.ca' + aTag.get('href')       # generate the job link from aTag
link = 'https://ca.indeed.com/viewjob?jk=f9c2351562959824&tk=1d1fc96bd40gp800&from=serp&vjs=3&advn=6904200094315584&adid=158881562&sjdu=wGm5HR2Y1KzsfK0G9eylwwedQmiyAygiQzR_6sI4tUZHSfIkRTBMOfhJoqWeQKp4wx0VeMwVL-UD7zZ7AP3Yl-_OFm6aH3hK-saRclqCL34RI5uxnhyfADD9Z_oR5elw1ZAYp0_z0qDLHuy_B1UgPX9aMle_Q6FgpZoZFKhK-Jtx14dQ3XWTn8XFpBtM-8Qg'
r = requests.get(link)                # open the job link
html_link = r.text                    # get the HTML from the job link
soup_link = BeautifulSoup(html_link, 'lxml')       # create soup from HTML
    
# extract job title from the job page
job_title = soup_link.find_all('h3', 
                                {'class':"icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title"})[0].text   
results[job_title] = {}  # create a sub-dict to store results for this job
# extract company name
results[job_title]['company_name'] = soup_link.find_all('div', {'class': 'icl-u-lg-mr--sm icl-u-xs-mr--xs'})[0].text 
# extract job description section
job_description = soup_link.find_all('div', {'class': 'jobsearch-JobComponent-description icl-u-xs-mt--md'})[0]
# extract minimum requirements -- 2nd instance of 'ul' in job page, split by '\n' separator into a list of str
results[job_title]['min_qualifications'] = job_description.select('ul:nth-of-type(2)')[0].text.split('\n')
results[job_title]['min_qualifications'].pop() # remove the last empty string from the list of qualifications
# extract preferred requirements -- 3rd instance of 'ul' in job page, split by '\n' separator into a list of strings
results[job_title]['preferred_qualifications'] = job_description.select('ul:nth-of-type(3)')[0].text.split('\n')
results[job_title]['preferred_qualifications'].pop() # remove the last empty string from the list of qualifications
# extract date posted
results[job_title]['date'] = soup_link.find_all('div', {'class':'jobsearch-JobMetadataFooter'})[0].text.split(' - ')[1]

results_df = pd.DataFrame(results)
results_df
```
